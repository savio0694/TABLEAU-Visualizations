### FEATURE ENGINEERING

```python
#divide the retail dataset into two, one from jan 2010 till september 2011 and another from 1 sep 2011 to 1 dec 2011
last_purchase_data=retail_data[(retail_data.InvoiceDate < pd.Timestamp(2011,9,1)) & (retail_data.InvoiceDate >= pd.Timestamp(2010,1,1))].reset_index(drop=True)
last_purchase_data.shape
```

```python
#1 sep 2011 to 1 dec 2011
next_purchase_data = retail_data[(retail_data.InvoiceDate < pd.Timestamp(2011,12,1)) & (retail_data.InvoiceDate >= pd.Timestamp(2011,9,1))].reset_index(drop=True)
next_purchase_data.shape
```

```python
#create a new dataframe
# Get the distinct customers from the last purchase dataframe
unique_customers = pd.DataFrame(last_purchase_data['Customer ID'].unique())


unique_customers.columns = ['CustomerID']
unique_customers
```

```python
# Create a dataframe with CustomerID and customers first purchase in first half

customer_1st_purchase_in_last_quarter = next_purchase_data.groupby('Customer ID').InvoiceDate.min().reset_index()
customer_1st_purchase_in_last_quarter.columns = ['CustomerID', 'MinPurchaseDate']

# Create a dataframe with CustomerID and customers last purchase in last 3 months

customer_last_purchase_in_first_half = last_purchase_data.groupby('Customer ID').InvoiceDate.max().reset_index()
customer_last_purchase_in_first_half.columns = ['CustomerID', 'MaxPurchaseDate']
```

```python
# Merge two dataframes of the last purchase and first half
customer_purchase_dates = pd.merge(customer_last_purchase_in_first_half,customer_1st_purchase_in_last_quarter, on='CustomerID', how='left')
```

```python
# Get the difference in days from MinPurchaseDate and MaxPurchaseDate for each customer
customer_purchase_dates['NextPurchaseDay'] = (customer_purchase_dates['MinPurchaseDate'] - customer_purchase_dates['MaxPurchaseDate']).dt.days
```

```python
# Update the dataframe unique_customers by merging it with the NextPurchaseDay column of the dataframe 
unique_customers = pd.merge(unique_customers, customer_purchase_dates[['CustomerID', 'NextPurchaseDay']], on='CustomerID', how='left')
# Fill all missing values in the dataset unique customers with the number 9999
unique_customers = unique_customers.fillna(9999)
unique_customers.head()
```

```python
# Get the maximum purchase date of each customer and create a dataframe with it together with the customer's id.
customer_max_purchase_date = retail_data.groupby('Customer ID').InvoiceDate.max().reset_index()
customer_max_purchase_date.columns = ['CustomerID','MaxPurchaseDate']

customer_max_purchase_date
```

### FREQUENCY


```python
# Get order counts for each user and create a dataframe with it
customer_frequency = retail_data.groupby('Customer ID').InvoiceDate.count().reset_index()
customer_frequency.columns = ['CustomerID', 'Frequency']


unique_customers = pd.merge(unique_customers, customer_frequency, on='CustomerID')
unique_customers=unique_customers[unique_customers['CustomerID']!="nan"]
```

```python
# Apply clustering to customer frequency data
kmeans = KMeans(n_clusters=3)
kmeans.fit(unique_customers[['Frequency']])
unique_customers['FrequencyCluster'] = kmeans.predict(unique_customers[['Frequency']])

# Order the dataframe based on the cluster values in FrequencyCluster
#uni = order_cluster(unique_customers, 'Frequency', 'FrequencyCluster', False)
unique_customers.head(30)
```

### REVENUE

```python
# Create a new label, Revenue of each item bought
retail_data['Revenue'] = retail_data.Price * retail_data.Quantity

# Get the revenue from each customer and sum them.
customer_revenue = retail_data.groupby('Customer ID').Revenue.sum().reset_index()
customer_revenue.columns = ['CustomerID', 'Revenue']

unique_customers = pd.merge(unique_customers, customer_revenue, on='CustomerID')
unique_customers.head()
```

```python
# Apply clustering to customer revenue data
kmeans = KMeans(n_clusters=4)
kmeans.fit(unique_customers[['Revenue']])
unique_customers['RevenueCluster'] = kmeans.predict(unique_customers[['Revenue']])

unique_customers.head()
```
